```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['company','name','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['company','name','fuel_type']),
                                       remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)*100
```

```python
scores=[]
for i in range(1000):
    # Correcting train_test_split to assign y_test properly
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=i)
    
    # Create the linear regression model and pipeline
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    
    # Fit the model on training data
    pipe.fit(x_train, y_train)
    
    # Make predictions on the test data
    y_pred = pipe.predict(x_test)
    
    # Print the RÂ² score for each iteration
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
print(np.argmax(scores))
```

```python
scores[np.argmax(scores)]*100
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred)*100
```
